```python
import random
import matplotlib.pyplot as plt
from collections import Counter

# Number of rolls
num_rolls = 10_000

# Rolling two dice
sums = [random.randint(1, 6) + random.randint(1, 6) for _ in range(num_rolls)]

# Counting frequency of each sum
freq = Counter(sums)

# Plotting the results
plt.bar(freq.keys(), freq.values(), color="blue", edgecolor="black")
plt.xlabel("Sum of Two Dice")
plt.ylabel("Frequency")
plt.title(f"Dice Roll Distribution ({num_rolls} Rolls)")

# Highlight sum = 7
plt.bar(7, freq[7], color="red", edgecolor="black", label="Thala - 7")
plt.legend()
plt.show()

# Printing actual probability
prob_7 = freq[7] / num_rolls
print(f"7 appeared {freq[7]} times out of {num_rolls} rolls ({prob_7:.2%} probability)")
```
