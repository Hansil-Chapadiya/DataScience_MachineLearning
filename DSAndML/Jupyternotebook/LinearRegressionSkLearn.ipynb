```python
# Import necessary libraries
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set plot size for better visualization
plt.rcParams['figure.figsize'] = (20.0, 10.0)

# Reading data from CSV file
data = pd.read_csv('headbrain.csv')
print(data.shape)  # Prints the shape of the dataset (rows, columns)

# Extracting independent (X) and dependent (Y) variables
X = data['Head Size(cm^3)'].values  # Head Size
Y = data['Brain Weight(grams)'].values  # Brain Weight

# Number of data points
m = len(X)

# Displaying first few rows of the data
data.head()

# Reshaping X for compatibility with scikit-learn (2D array required)
X = X.reshape((m, 1))

# Creating a Linear Regression model
reg = LinearRegression()

# Fitting the model to the data
reg = reg.fit(X, Y)

# Predicting Y values using the trained model
Y_pred = reg.predict(X)

# Calculating Mean Squared Error (MSE)
mse = mean_squared_error(Y, Y_pred)

# Calculating Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Calculating R-squared value
r2_score = reg.score(X, Y)

# Printing RMSE and R-squared values
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R-squared Score: {r2_score}")

# Plotting the data and regression line
plt.scatter(X, Y, color='#ef5423', label='Scatter Plot')  # Scatter plot of actual data
plt.plot(X, Y_pred, color='#58b970', label='Regression Line')  # Regression line
plt.xlabel('Head Size (cm³)')
plt.ylabel('Brain Weight (grams)')
plt.legend()
plt.show()
```
