```python
# Importing necessary libraries
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Setting plot size for better visualization
plt.rcParams['figure.figsize'] = (20.0, 10.0)

# Reading data from CSV file
data = pd.read_csv('headbrain.csv')
print(data.shape)  # Prints the shape of the dataset (rows, columns)
data.head()  # Displays the first 5 rows of the dataset

# Extracting independent (X) and dependent (Y) variables
X = data['Head Size(cm^3)'].values  # Head Size
Y = data['Brain Weight(grams)'].values  # Brain Weight

# Calculating means of X and Y
mean_x = np.mean(X)
mean_y = np.mean(Y)

# Calculating the coefficients of the regression line (b1 and b0)
# Formula for slope (b1) and intercept (b0):
# b1 = ((Xi - mean_X) * (Yi - mean_Y)) / ((Xi - mean_X)^2)
# b0 = mean_Y - b1 * mean_X
n = len(X)  # Total number of values
numerator = 0  # Numerator for b1 calculation
denominator = 0  # Denominator for b1 calculation

# Loop to calculate numerator and denominator
for i in range(n):
    numerator += (X[i] - mean_x) * (Y[i] - mean_y)
    denominator += (X[i] - mean_x) ** 2

# Slope (b1) and Intercept (b0)
b1 = numerator / denominator
b0 = mean_y - (b1 * mean_x)

# Printing the calculated coefficients
print(f"Slope (b1): {b1}, Intercept (b0): {b0}")

# Plotting the regression line
# Define the range for the line using min and max of X, with some buffer
max_x = np.max(X) + 100
min_x = np.min(X) - 100

# Generate 1000 evenly spaced points in the range [min_x, max_x]
x = np.linspace(min_x, max_x, 1000)

# Calculate corresponding y values using the regression equation
y = b0 + b1 * x  # y = b0 + b1*x

# Plot the regression line
plt.plot(x, y, color='#58b970', label='Regression Line')

# Plot scatter points for actual data
plt.scatter(X, Y, c='#ef5423', label='Scatter Plot')

# Adding labels and legend
plt.xlabel('Head Size (cm)')
plt.ylabel('Brain Weight (grams)')
plt.legend()

# Display the plot
plt.show()

# Calculating R-squared (coefficient of determination)
# SS_total (total sum of squares) and SS_residual (residual sum of squares)
ss_t = 0  # Total sum of squares
ss_r = 0  # Residual sum of squares
for i in range(n):
    y_pred = b0 + b1 * X[i]  # Predicted value using the regression equation
    ss_t += (Y[i] - mean_y) ** 2  # Total variation
    ss_r += (Y[i] - y_pred) ** 2  # Residual variation

# R-squared formula: R = 1 - (SS_residual / SS_total)
r2 = 1 - (ss_r / ss_t)

# Printing R-squared value
print(f"R-squared value: {r2}")
```
