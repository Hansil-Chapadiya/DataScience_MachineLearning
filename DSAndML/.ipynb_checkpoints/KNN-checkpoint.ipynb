```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from matplotlib.colors import ListedColormap

# Step 2: Create Sample Dataset (Height, Weight, and Category)
data = {
    "Height": [150, 160, 170, 180, 175, 165, 155, 185, 172, 158],
    "Weight": [50, 60, 70, 80, 75, 65, 55, 85, 72, 57],
    "Category": ["Light", "Light", "Medium", "Heavy", "Heavy", "Medium", "Light", "Heavy", "Medium", "Light"]
}

df = pd.DataFrame(data)

# Step 3: Convert Categorical Target Variable into Numeric
df["Category"] = df["Category"].map({"Light": 0, "Medium": 1, "Heavy": 2})

# Step 4: Split Data into Features and Target
X = df[["Height", "Weight"]].values  # Features (Height and Weight)
y = df["Category"].values  # Target (Weightlifting Category)

# Step 5: Train KNN Model
knn = KNeighborsClassifier(n_neighbors=3)  # Using k=3
knn.fit(X, y)

# Step 6: Predict Category for Given Data (Height = 161, Weight = 61)
test_data = np.array([[161, 61]])
predicted_category = knn.predict(test_data)

# Step 7: Convert Prediction Back to Category Name
category_mapping = {0: "Light", 1: "Medium", 2: "Heavy"}
print(f"Predicted Category for Height 161 cm & Weight 61 kg: {category_mapping[predicted_category[0]]}")

# Step 8: Plot Decision Boundary
h = 1  # Step size in the mesh
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
cmap_light = ListedColormap(["#FFAAAA", "#AAFFAA", "#AAAAFF"])
cmap_bold = ["red", "green", "blue"]
plt.contourf(xx, yy, Z, cmap=cmap_light, alpha=0.3)

# Scatter plot of actual points
for i, color in enumerate(cmap_bold):
    plt.scatter(X[y == i, 0], X[y == i, 1], c=color, label=category_mapping[i], edgecolor='k', s=100)

# Scatter plot for test point
plt.scatter(test_data[0][0], test_data[0][1], c="yellow", edgecolor='black', s=200, marker='*', label="Test Point")

plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("KNN Classification - Weightlifting Category")
plt.legend()
plt.show()
```
